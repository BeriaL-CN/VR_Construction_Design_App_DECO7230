```python
# 根据数据画一个柱图表
# x轴是参与者名字，y轴是得分
# interaction1
import matplotlib.pyplot as plt
import numpy as np

peopleNames = ('1', '2', '3', '4')
efficiency = (5, 5, 5, 2)
intuitive = (5, 5, 4, 5)

fig, ax = plt.subplots()
index = np.arange(len(peopleNames))
bar_width = 0.35

rects1 = ax.bar(index, efficiency, bar_width, label='Efficiency')
rects2 = ax.bar(index + bar_width, intuitive, bar_width, label='Intuitive')

ax.set_xlabel('People')
ax.set_ylabel('Scores')
ax.set_title('Interaction1')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(peopleNames)
ax.legend()

fig.tight_layout()
plt.show()
```
