```python
# 根据数据画一个柱图表
# x轴是参与者名字，y轴是得分
# interaction1
import matplotlib.pyplot as plt
import numpy as np

peopleNames = ('1', '2', '3', '4')
intuitive = (5, 4, 4, 4)
easy_to_learn = (5, 5, 3, 5)
useful = (5, 5, 4, 5)

fig, ax = plt.subplots()
index = np.arange(len(peopleNames))
bar_width = 0.15

rects1 = ax.bar(index, intuitive, bar_width, label='Intuitive')
rects2 = ax.bar(index + bar_width, easy_to_learn, bar_width, label='Easy to Learn')
rects3 = ax.bar(index + bar_width * 2, useful, bar_width, label='Useful')

ax.set_xlabel('People')
ax.set_ylabel('Scores')
ax.set_title('Zooming Interaction')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(peopleNames)
ax.legend()

fig.tight_layout()
plt.show()
```
